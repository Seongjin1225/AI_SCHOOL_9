```python
import pandas as pd
import matplotlib.pyplot as plt
pitcher = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/데이터 분석/picher_stats_2017.csv')
batter = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/데이터 분석/batter_stats_2017.csv')

pitcher.head()
batter.head()

pitcher.rename(columns={'연봉(2018)':'2018년', '연봉(2017)':'2017년'},inplace=True)
batter.rename(columns={'연봉(2018)':'2018년', '연봉(2017)':'2017년'},inplace=True)
```

```python
# 투수 연봉 top10(2017년)
wage_2017 = pitcher[['선수명','2017년']]
top_2017 = wage_2017.sort_values(by='2017년',ascending=False)
top_2017.set_index('선수명',inplace=True)

plt.rc('font', family='NanumBarunGothic')
top_2017[:10].plot.bar(rot=45)
plt.title('2017년 투수 연봉 순위 top10')
```

```python
# 투수 연봉 top10(2018년)
wage_2018 = pitcher[['선수명','2018년']]
top_2018 = wage_2018.sort_values(by='2018년',ascending=False)
top_2018.set_index('선수명',inplace=True)
plt.rc('font', family='NanumGothic')
top_2018[:10].plot.bar(rot=45)
plt.title('2018년 투수 연봉 순위 top10')
```

```python
# 타자 연봉 top10(2017년)
b_wage_2017 = batter[['선수명','2017년']]
b_top_2017 = b_wage_2017.sort_values(by='2017년',ascending=False)
b_top_2017.set_index('선수명',inplace=True)

plt.rc('font', family='NanumGothic')
b_top_2017[:10].plot.bar(rot=35)
plt.title('2017년 타자 연봉 순위 top10')
```

```python
# 타자 연봉 top10(2018년)
b_wage_2018 = batter[['선수명','2018년']]
b_top_2018 = b_wage_2018.sort_values(by='2018년',ascending=False)
b_top_2018.set_index('선수명',inplace=True)

plt.rc('font', family='NanumGothic')
b_top_2017[:10].plot.bar(rot=35)
plt.title('2018년 타자 연봉 순위 top10')
```

```python
# 연봉과 출루율의 상관관계
import seaborn as sns
data = batter[['출루율','2017년','2018년']]
data.head()

corr_1 = data.corr()
sns.heatmap(corr_1,annot=True,fmt='.2f')
```

```python
# 연봉과 era의 상관관계
data = pitcher[['ERA','2017년','2018년']]
corr_d = data.corr()
sns.heatmap(corr_d, annot=True,fmt='.2f')
```

```python
# 연봉과 지표의 상관관계(투수 ver.)
import numpy as np
data = pitcher[['BABIP', 'LOB%', 'ERA', 'RA9-WAR','FIP', 'kFIP', 'WAR', '2017년', '2018년']]
corr_d = data.corr()

upp_mat = np.triu(corr_d)
sns.heatmap(corr_d,annot=True,fmt='.2f',mask=upp_mat)
```

```python
# 연봉과 지표의 상관관계(타자 ver.)
data = batter[['BABIP', '타율', '출루율', '장타율',	'OPS', 'wOBA', 'WAR', '2017년', '2018년']]
corr_d = data.corr()

upp_mat = np.triu(corr_d)
sns.heatmap(corr_d,annot=True,fmt='.2f',mask=upp_mat)
```

```python
# 2017년과 2018년 연봉 차익 계산해서 diff 컬럼 만들기
# 투수
pitcher['difference'] = pitcher['2018년'] - pitcher['2017년']
pitcher.head()

# 타자
batter['difference'] = batter['2018년'] - batter['2017년']
batter.head()

pitcher.head()
```

```python
# 연봉 변화와 지표의 상관관계
data = pitcher[['BABIP', 'LOB%', 'ERA', 'RA9-WAR','FIP', 'kFIP', 'WAR','difference']]
corr_d = data.corr()

# upp_mat = np.triu(corr_d)
sns.heatmap(data,annot=True,fmt='.2f')
```

```python
# 가장 많은 이닝을 소화한 투수
data = pitcher[['선수명','이닝']]
data = data.set_index('선수명')
data.sort_values(by='이닝',ascending=False)
data[:10].plot.bar(rot=45)
```

```python
# 팀별 승리 횟수
data = pitcher.groupby('팀명')['승'].size()

data.plot.bar(rot=45)
plt.title('2017년 팀별 승수')
```

```python
total_mean = pitcher['2017년'].mean()
wage_mean = pitcher.groupby('팀명')['2017년'].mean()

money = wage_mean.index.tolist()
money.append('mean')
pos = np.arange(len(money))
win = wage_mean.tolist()
win.append(total_mean)

bar_list = plt.bar(pos,win,align='center',alpha=0.5)
bar_list[-4].set_color('r')

plt.axhline(y=total_mean, color='b', linestyle='--')
plt.xticks(pos, money)

plt.title('2017년 구단별 투수 평균 연봉')
plt.show()

wage_mean
```
