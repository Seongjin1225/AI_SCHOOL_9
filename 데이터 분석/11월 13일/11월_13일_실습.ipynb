```python
# 제너레이터는 iterator 객체의 한 종류이다
# 함수를 기반으로 제너레이터를 만든 예는 다음과 같다
# 함수 몸체에서 yield가 하나라도 보이면 단순한 함수의 정의가 아닌
# 제너레이터 함수의 정의가 된다
# 제너레이터의 정체 : 함수의 몸체 부분을 실행하는 것
def gen_num():
    print('first number')
    yield 1  # yield 보이면 무조건 generator
    print('second number')
    yield 2
    print('third number')
    yield 3

gen = gen_num()

next(gen) # 첫번째 yield문을 만날때 까지 실행 이어감
next(gen)
next(gen)
next(gen) # 함수 호출 이후에
```

# 자동차 연비 분석

```python
import numpy as np
import pandas as pd
data_df = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/auto-mpg.csv', header=0, engine='python')
print('데이터셋 크기 : ', data_df.shape)
data_df.head()
```

```python
data_df.drop(['car_name', 'origin','horsepower'], axis=1, inplace=False)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# X,Y 분할하기  - 종속, 독립변수 전처
Y = data_df['mpg']
X = data_df.drop(['mpg'], axis=1, inplace=False)
```

```python
# 훈련용 데이터와 평가용 데이터 분할하기
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
# 선형회귀분석 : 모델 생성
lr = LinearRegression()
```

```python
# 선형회귀분석 : 모델 학습
lr.fit(X_train, Y_train)
```

```python
# 선형회귀분석 : 평가 데이터에 대한 예측
Y_predict = lr.predict(X_test)
```

```python
# mse - 평가지표, rmse - 오차 제곱근법
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse,rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편 값: ', np.round(lr.intercept_, 2))
print('회귀 계수 값: ', np.round(lr.coef_, 2))
```

```python
coef = pd.Series(data=np.round(lr.coef_, 2), index=X.columns)
coef.sort_values(ascending=False)
```
