```python
import pandas as pd
cereal = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/cereals.csv')
cereal.head()
```

```python
# 각 제조업체의 Fiber 열에서 최소값을 찾으세요.
data.Fiber.min()
```

```python
# 제조업체별로 설탕함량이 가장 낮은 시리얼을 새로운 Dataframe으로 추출하세요.
def smallest_sugar_row(df):
    return df.nsmallest(1, 'Sugars')
data.apply(smallest_sugar_row)
```

# txt 활용 자연어 처리

```python
import pandas as pd
import matplotlib.pyplot as plt
from konlpy.tag import Okt
from wordcloud import WordCloud
okt = Okt()

df = pd.read_table('/content/drive/MyDrive/AI스쿨 파일/ratings_train.txt')
df
```

```python
# 중복값, 결측치 확인 및 제거
# 중복값 제거 --> 필수 ㄴㄴ 필요에 따라 안할수도 ㅇㅇ

df.isnull().sum()
df['id'].nunique()

df = df.dropna(how='any')# Null 값이 존재하는 행 제거
df.isnull().sum()
```

```python
# 불용어 제거
df['document'] = df['document'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','')
```

```python
temp_list = []

for sentence in df['document']:
    s_list = okt.pos(sentence)
    for word, tag in s_list:
        if tag in ['Noun', 'Adjective']:
            temp_list.append(word)
counts = collections.Counter(temp_list)
tag = counts.most_common(10)
tag
```

```python
wc = WordCloud(font_path = 'NanumGothic', background_color='skyblue', max_font_size=60)
cloud = wc.generate_from_frequencies(dict(tag))

plt.figure(figsize=(10,8))
plt.imshow(cloud)  # 워드클라우드 나타낼때는 imshow()
```

# 오후 진도
## 자연어 처리 실습(영화 리뷰 분석)

```python
# 랭킹이 가장 높은 영화 워드클라우드로 표시
import pandas as pd
import matplotlib.pyplot as plt
import collections
from konlpy.tag import Okt
from wordcloud import WordCloud
okt = Okt()
df = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/comment_rank.csv')
df.head()

# 결측치 제거
df.isnull().sum()
df = df.dropna(how='any')
df.isnull().sum()
```

```python
# movie로 그룹화 한다음 rank를 sum해서 가장 큰 값으로 ㄱㄱ
data = df.groupby('movie')['rank'].sum()
data.nlargest(5)
```

```python
# 고양이 집사 영화 코멘트 중 가장 많이 쓰인 단어 10개
data = df[df.movie=='고양이 집사']
data['comment'] = data['comment'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")

temp_list = []
for sentence in data['comment']:
    s_list = okt.pos(sentence)
    for word, tag in s_list:
        if tag in ['Noun', 'Adjective']:
            temp_list.append(word)
counts = collections.Counter(temp_list)
tag = counts.most_common(10)
tag

wc = WordCloud(font_path='NanumGothic', background_color='blue', max_font_size=70)
cloud = wc.generate_from_frequencies(dict(tag))

plt.figure(figsize=(10,8))
plt.imshow(cloud)
```
