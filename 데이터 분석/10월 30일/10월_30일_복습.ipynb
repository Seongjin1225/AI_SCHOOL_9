```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel('/content/drive/MyDrive/AI스쿨 파일/데이터 분석/health_screenings_2020_1000ea.xlsx')

# 키 및 몸무게 산점도 그래프
heightdata = data['height']
weightdata = data['weight']

plt.figure(figsize=(10,4))
plt.scatter(heightdata,weightdata)
plt.xlabel('height')
plt.ylabel('weight')
plt.title('2020 Health Screenings Scatter Graph')
plt.grid()
plt.show()
```

```python
# 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래프
# 각각의 관계를 하나씩 나타내는 방식으로 해야함
systolic_data = data['systolic']
diastolic_data = data['diastolic']
blood_sugar_data = data['blood_sugar']

plt.figure(figsize=(10,6))

plt.scatter(systolic_data, diastolic_data, color = 'r', edgecolor='w', label='systolic*diastolic')
plt.scatter(systolic_data, blood_sugar_data, color = 'g', edgecolor='w', label='systolic*blood_sugar')
plt.scatter(diastolic_data, blood_sugar_data, color = 'b', edgecolor='w', label='diastolic*blood_sugar')

plt.xlim(40,180)
plt.ylim(30,300)
plt.title('2020 Health Screenings Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# seaborn part
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('/content/drive/MyDrive/AI스쿨 파일/health_screenings_2020_1000ea.xlsx')

# 성별, 음주 여부, 흡연 상태 데이터 전처리
health = data.loc[:,['gender','height','weight','waist','drinking','smoking']]

health.loc[health['gender']==1,['gender']]='Male'
health.loc[health['gender']==2,['gender']]='Female'
health.loc[health['drinking']==0, ['drinking']] = 'Non-drinking'
health.loc[(health['drinking']==1)|(health['smoking']==2), ['smoking']] = 'Non-smoking'
health.loc[health['smoking']==3, ['smoking']] = 'Smoking'

# 음주 여부 및 흡연 상태 데이터 준비하기
drinking = health.groupby(['gender','drinking'])['drinking'].count()
smoking = health.groupby(['gender','smoking'])['smoking'].count()
drinking = drinking.to_frame(name='count')
smoking = smoking.to_frame(name='count')
drinking = drinking.reset_index()
smoking = smoking.reset_index()

# 시본 막대 그래프
fig = plt.figure(figsize=(17,6))
fig.suptitle('2020 Health Screenings Drinking & Smoking Type Bar Graph', fontweight='bold')
index = np.arange(4)

fig.add_subplot(1,2,1)
plt.bar(index, drinking['count'])
plt.title('Drinking Type')
plt.ylabel('Count')
plt.xticks(index, ['Non-drinking(Female)', 'Drinking(Female)', 'Non-drinking(Male)', 'Drinking(Male)'])

fig.add_subplot(1,2,2)
plt.bar(index, smoking['count'])
plt.title('Smoking Type')
plt.ylabel('Count')
plt.xticks(index, ['Non-smoking(Female)', 'Smoking(Female)', 'Non-smoking(Male)', 'Smoking(Male)'])

plt.show()
```

```python
fig = plt.figure(figsize=(17,6))

area1 = fig.add_subplot(1,2,1)
area2 = fig.add_subplot(1,2,2)

ax1 = sns.barplot(data=drinking, x='gender', y='count', hue='drinking', ax=area1)  # hue -> legend와 비슷, 더 많은 정보 표현 가
ax2 = sns.barplot(data=smoking, x='gender', y='count', hue='smoking', ax=area2)

fig.suptitle('2020 Health Screenings Drinking & Smoking Type Bar Graph', fontweight='bold')
area1.set_title('Drinking Type')
area2.set_title('Smoking Type')

plt.show()
```

```python
# 병원 환자 노쇼 데이터 전처리 실습
import numpy as no
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 전처리 시작에서 하는 공식과 같은 과정
medi = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/medical.csv')  # 읽어오기
medi.head()  # 정보 확인
medi.columns  # 컬럼 확인
medi.info()  # 기본 정보 확인
medi.isnull().sum()  # 결측치 확인
```

```python
medi.describe()
medi = medi[medi['Age']>=0]
medi.head()
```

```python
# no-show의 데이터 타입 문자형 -> 수치형으로
medi['No-show'] = medi['No-show'].map({'No':0,'Yes':1})
medi['No-show'].value_counts()
```

```python
# 예약날짜, 방문 날짜 자료형을 날짜형으로 대체
medi['ScheduledDay'] = pd.to_datetime(medi['ScheduledDay'])
medi['AppointmentDay'] = pd.to_datetime(medi['AppointmentDay'])
medi.dtypes
```

```python
# 예약한 날 - 실제 방문날 계산하여 waiting_day 컬럼 만들어 추가
# 일수차이 구하기 위해 dayofyear 사용
medi['waiting_day'] = medi['AppointmentDay'].dt.dayofyear - medi['ScheduledDay'].dt.dayofyear
medi.head()
```
