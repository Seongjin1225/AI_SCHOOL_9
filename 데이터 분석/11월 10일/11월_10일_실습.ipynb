# 회귀분석
- 와인 품질 등급 예측
- 기술 통계, 회귀 분석, t-검정, 히스토그램
- 보통 맨 마지막 컬럼 -> 종속변수(y)

```python
import pandas as pd
red_df = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/winequality-red2.csv')
white_df = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/winequality-white2.csv')

red_df.to_csv('/content/drive/MyDrive/AI스쿨 파일/winequality-red2.csv', index = False)
white_df.to_csv('/content/drive/MyDrive/AI스쿨 파일/winequality-white2.csv', index = False)
red_df.head()

# type 컬림 추가
red_df.insert(0, column = 'type', value = 'red')
red_df.shape

white_df.head()
white_df.insert(0, column = 'type', value = 'white')
white_df.head()

# 합치기(옆으로)
wine = pd.concat([red_df, white_df])
wine.shape
wine.to_csv('/content/drive/MyDrive/AI스쿨 파일/wine.csv', index = False)
wine.columns = wine.columns.str.replace(' ', '_')
wine.head()
wine.describe
```

```python
sorted(wine.quality.unique())
wine.quality.value_counts()
```

```python
from scipy import stats
from statsmodels.formula.api import ols, glm

red_wine_quality = wine.loc[wine['type']=='red', 'quality']
white_wine_quality = wine.loc[wine['type']=='white', 'quality']

# t-검정(그룹 간의 차이를 확인)
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
```

```python
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + \
            residual_sugar + chlorides + free_sulfur_dioxide + \
            total_sulfur_dioxide + density + pH + sulphates + alcohol'
# 위 코드는 종속변수(quality) + 독립변수 형태로 되어있음
# ols라는 회귀분석 툴 쓰려면 위 Rformula 형태 지켜야함 ㅇㅇ

regression_result = ols(Rformula, data=wine).fit()
regression_result.summary()
```

```python
# 테스트 해보기 위해서 두 컬럼 제외함
sample1 = wine[wine.columns.difference(['quality','type'])]  # 괄호 안 컬럼 제외하고 출력
sample1 = sample1[0:5][:]

sample1_predict = regression_result.predict(sample1)
sample1_predict
```
