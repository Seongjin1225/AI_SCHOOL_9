## 서울 지하철 데이터 전처리

```python
# 라이브러리 탑재
import folium
import pandas as pd

# 기본 서울 지도 그래프 표기
seoul = folium.Map(location=[37.55, 126.98], zoom_start = 12)
seoul

# 승하차 인원 파일 읽어오기
data = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/seoul-metro-2021.logs.csv')

# 지하철 역 정보 파일 읽어오기
station_info = pd.read_csv('/content/drive/MyDrive/AI스쿨 파일/seoul-metro-station-info.csv')
# station_info.head()

# 승하차 인원 데이터 상세 정보 확인
# data.info()

# 역 기준으로 승하차 인원 합치기
station_sum = data.groupby("station_code").sum(numeric_only=False)
station_sum
# 지하철 역 데이터 상세정보 확인
#  station_info.info()

# 필요한 컬럼만 가져오기(역 코드, 위도, 경도)
station_info = station_info[['station.code','geo.latitude','geo.longitude']]
# station_info

# # 승하차 인원 데이터와 색인 맞추기
station_info = station_info.set_index('station.code')
station_info.head()

# # # # 두 데이터 합치기(join 사용)
joined_data = station_sum.join(station_info)
joined_data
```

```python
# 지도 부분
# 승차용 서울 지도 만들기
seoul_in = folium.Map(location=[37.55, 126.98], zoom_start=12)

# 지도 확인
seoul_in
```

```python
# 히트맵 플러그인 모듈 탑재
from folium.plugins import HeatMap

# 지도 추가하기
HeatMap(data=joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(seoul_in)

seoul_in
```

```python
# 하차용 서울 지도 만들기
seoul_out = folium.Map(location=[37.55, 126.98], zoom_start=12)

# 지도 추가하기
HeatMap(data=joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(seoul_out)

seoul_out
```

```python
# 출퇴근 데이터 추출 + 합치기
# 출근
morning_data = data[pd.to_datetime(data.timestamp).dt.hour < 9]
morning_data

# 퇴근
evening_data = data[pd.to_datetime(data.timestamp).dt.hour > 17]
evening_data
```

```python
# 역 기준 그룹화하여 데이터
morning_station_sum = morning_data.groupby('station_code').sum(numeric_only=False)
evening_station_sum = evening_data.groupby('station_code').sum(numeric_only=False)

# 합치기
morning_joined_data = morning_station_sum.join(station_info)
evening_joined_data = evening_station_sum.join(station_info)
```

```python
# 출근시간 승차 인원
morning_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(morning_seoul_in)
morning_seoul_in
```

```python
# 출근시간 하차 인원
morning_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(morning_seoul_out)
morning_seoul_out
```

```python
# 문제 1.
# 점심시간 승하차 현황 분석(11시~2시)
# 강남, 강북 지역에서 승차 인원이 가장 많은 역 3개
# 위와 동일 but 하차 인원

lunch_data = data[(pd.to_datetime(data.timestamp).dt.hour <= 14) & (pd.to_datetime(data.timestamp).dt.hour >=11)]
lunch = lunch_data.groupby('station_code').sum(numeric_only=False)
lunch_join = lunch.join(station_info)

# 승차 인원 확인
lunch_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = lunch_join[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(lunch_seoul_in)
lunch_seoul_in
```

```python
# 하차 인원 확인
lunch_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = lunch_join[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(lunch_seoul_out)
lunch_seoul_out
```

```python
# 문제 2
# 문제 1에서 가공한 데이터 가지고 히트맵 대신 원하는 역에 동그라미 ㅇㅇ
# 조건
# --> 점심시간 하차인원 250만 이상인 역
#     동그라미 테두리는 붉은색, 내부는 붉은색 음영
#     반지름은 하차인원의 0.01%
lunch_seoul_out_circle = folium.Map(location=[37.55, 126.98], zoom_start = 12)

for index, row in lunch_join[['geo.latitude', 'geo.longitude', 'people_out']].iterrows():  #  iterrows() 함수는 데이터 프레임의 각 행에 대한 인덱스(index)와 해당 행(row)을 반환
    if int(row['people_out']) > 2_500_000:
        folium.Circle(location=[str(row['geo.latitude']), str(row['geo.longitude'])],   color='Red', fill=True,
              radius=(int(row['people_out'])/10000)).add_to(lunch_seoul_out_circle)

lunch_seoul_out_circle
```
